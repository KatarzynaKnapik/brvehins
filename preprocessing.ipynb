```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns

import pyreadr
```

```python
brvehins1_ls = ["brvehins1a.rda", "brvehins1b.rda", "brvehins1c.rda", "brvehins1d.rda", "brvehins1e.rda"]
df_list=[]
for dataset in brvehins1_ls:
    data_dict = pyreadr.read_r("./dane/" + dataset)
    print(data_dict)
    df = data_dict[dataset.split(".")[0]]
    df_list.append(df)

df = pd.concat(df_list, ignore_index=True)
```
